```python
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Enhance edges and contrast
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blur, 30, 150)

    # Use dilation + closing to connect components better
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
    dilated = cv2.dilate(edged, kernel, iterations=2)
    closed = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel)

    return img, gray, closed
```

```python
def detect_components(thresh_img):
    contours, _ = cv2.findContours(thresh_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    components = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        if w > 5 and h > 5:  # Detect smaller elements
            components.append((x, y, w, h))

    return components
```

```python
import pytesseract

def extract_text(img, box):
    x, y, w, h = box
    roi = img[y:y+h, x:x+w]
    text = pytesseract.image_to_string(roi, config='--psm 6')
    return text.strip().lower()
```

```python
def compute_similarity(base_count, matched_count):
    if base_count == 0:
        return 0
    return round((matched_count / base_count) * 100, 2)
```

```python
def draw_annotations(img, missing, misplaced, overlapped):
    print("Overlapped components:", overlapped)  # Debugging line

    # Ensure each component has the correct format
    overlapped = [box for box in overlapped if len(box) == 4]

    for (x, y, w, h) in missing:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)  # Red for missing
        cv2.putText(img, 'Missing', (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

    for (x, y, w, h) in misplaced:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 255), 2)  # Yellow for misplaced
        cv2.putText(img, 'Misplaced', (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 1)

    for (x, y, w, h) in overlapped:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Blue for overlapped
        cv2.putText(img, 'Overlapped', (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)




    return img
```

```python
from skimage.metrics import structural_similarity as ssim
from matplotlib import pyplot as plt
import cv2

def run_ui_compare(base_path, test_path):
    base_img, _, base_thresh = preprocess_image(base_path)
    test_img, _, test_thresh = preprocess_image(test_path)

    base_comps = detect_components(base_thresh)
    test_comps = detect_components(test_thresh)

    print(f"Base components detected: {len(base_comps)}")
    print(f"Test components detected: {len(test_comps)}")

    matched, misplaced, missing, overlapped = compare_components(base_comps, test_comps)

    similarity = compute_similarity(len(base_comps), len(matched))

    # For annotation
    annotated_img = draw_annotations(test_img.copy(), missing, misplaced, overlapped)

    # SSIM comparison
    base_gray = cv2.cvtColor(base_img, cv2.COLOR_BGR2GRAY)
    test_gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
    test_resized = cv2.resize(test_gray, (base_gray.shape[1], base_gray.shape[0]))
    ssim_score, _ = ssim(base_gray, test_resized, full=True)

    # Print stats
    print(f"Component-Based Similarity Score: {similarity}%")
    print(f"SSIM Visual Similarity: {round(ssim_score * 100, 2)}%")
    print(f"Missing Components: {len(missing)}")
    print(f"Misplaced Components: {len(misplaced)}")
    print(f"Overlapped Pairs: {len(overlapped)}")

    # Convert images to RGB for matplotlib
    base_rgb = cv2.cvtColor(base_img, cv2.COLOR_BGR2RGB)
    test_rgb = cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB)
    annotated_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)

    # Display all three images side by side
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(base_rgb)
    plt.title("Base Image")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(test_rgb)
    plt.title("Test Image")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(annotated_rgb)
    plt.title("Differences Annotated")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
    print("Drawing overlapped boxes...")
    for box in overlapped:
      print("Box:", box)

    return annotated_img
```

```python
from matplotlib import pyplot as plt

def visualize_detected_components(image_path):
    img, _, thresh = preprocess_image(image_path)
    comps = detect_components(thresh)

    for x, y, w, h in comps:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 1)
        cv2.putText(img, 'COMP', (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 1)

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.title("Detected UI Components")
    plt.axis("off")
    plt.show()

    print(f"Components detected: {len(comps)}")
```
